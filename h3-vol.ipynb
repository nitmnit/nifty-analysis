```python
import pandas as pd
pd.options.display.max_colwidth = 10
pd.options.display.max_rows = None
```

```python
candles = pd.read_json("tv-22400-call.json")
candles['Time'] = pd.to_datetime(candles['Time'])
candles.set_index('Time', inplace=True)
candles.sort_index(inplace=True)
R = 'red'
G = 'green'
candles["color"] = candles.apply(lambda r: G if r.open < r.close else R, axis=1)
candles["last_red_open"] = pd.NA

cur_last_red_open = pd.NA
for idx, row in candles.iterrows():
    if row.color == R:
        cur_last_red_open = row.open
    else:
        candles.at[idx, "last_red_open"] = cur_last_red_open
```

```python
candles["above_red"] = candles[candles.last_red_open.notna()].open > candles[candles.last_red_open.notna()].last_red_open
```

```python
from bokeh.plotting import figure, show
from bokeh.io import show, output_notebook
output_notebook()

def bokeh_plot_ts(df):
    TOOLS = "pan,wheel_zoom,box_zoom,reset,save"
    p = figure(tools=TOOLS, width=1500, height=800, title="Candles", background_fill_color="black")
    p.xaxis.major_label_orientation = 0.8 # radians
    p.x_range.range_padding = 0.05
    
    # map dataframe indices to date strings and use as label overrides
    # p.xaxis.major_label_overrides = {
    #     i: i.strftime('%H:%M:%S') for i in df.index.to_pydatetime()
    # }
    p.xaxis.major_label_overrides = {
        i: f'{timestamp.hour}:{timestamp.minute}:{timestamp.second}' for i, timestamp in enumerate(df.index)
    }
    p.xaxis.ticker.desired_num_ticks = 10

    # one tick per week (5 weekdays)
    # p.xaxis.ticker = list(range(df.index.to_pydatetime()[0], df.index[-1], 5))
    
    p.segment(df.index, df.high, df.index, df.low, color="white")
    inc = df.close > df.open
    dec = df.open > df.close
    dec_ilocs = list(range(0, df[dec].shape[0]))
    inc_ilocs = list(range(0, df[inc].shape[0]))
    p.vbar(dec_ilocs, pd.Timedelta(seconds=5), df.open[dec], df.close[dec], fill_color="#EF5451", line_color="#EF5451", line_width=1, width=.5)
    p.vbar(inc_ilocs, pd.Timedelta(seconds=5), df.open[inc], df.close[inc], fill_color="#27A69B", line_color="#27A69B", line_width=1, width=.5)
    show(p)

bokeh_plot_ts(candles)
# print(dir(candles.iloc[0].name))
# print(type(candles.iloc[0].name.to_pydatetime()))
# # print(candles.iloc[0].name.second)
ddfasdf = {
        i: f'{timestamp.hour}:{timestamp.minute}:{timestamp.second}' for i, timestamp in enumerate(candles.index)
    }

# print(len(ddfasdf))
# print(len(candles))
print(ddfasdf)
```
