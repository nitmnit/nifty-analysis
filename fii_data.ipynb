```python
import datetime as dt
import os
from files_download import fetch_nifty_futures_data

import pandas as pd
import holoviews as hv
from bokeh.plotting import show
from bokeh.models import NumeralTickFormatter
hv.extension('bokeh')


end_date = dt.datetime.now()
start_date = end_date - dt.timedelta(days=365)
folder_path = os.path.join(os.getcwd(), "data", "fii")

futures_data, oi_data = fetch_nifty_futures_data(start_date, end_date, folder_path)
# print(futures_data)
# print(oi_data)
```

```python
import pandas as pd
import holoviews as hv
from bokeh.plotting import show
from bokeh.models import NumeralTickFormatter, DatetimeTickFormatter, HoverTool
hv.extension('bokeh')

# Your code to calculate net_futures
oi_data.loc[:, "net_futures"] = oi_data.future_index_long - oi_data.future_index_short

# Group the data by client type and create separate curves for each type
curves = []
for client_type, group in oi_data.loc[(oi_data.client_type == "FII") | (oi_data.client_type == "Client")].groupby('client_type'):
    group = group.reset_index()
    curve = hv.Curve(group, 'date', 'net_futures', label=f'Net Futures - {client_type}')
    curves.append(curve)

# Overlay all the curves
net_futures_plot = hv.Overlay(curves)

# Add HoverTool to display date and net_futures value
hover = HoverTool(
    tooltips=[
        ("Date", "@date{%d %b %Y}"),  # Display the month and year
        ("Net Futures", "@net_futures{0,0}")  # Display the net_futures value with commas
    ],
    formatters={
        '@date': 'datetime'  # Use 'datetime' formatter for the date
    },
    mode='vline'  # Display a vertical line across the plot at the hovered point
)

# Hook function to manually add the HoverTool
def add_hover_tool(plot, element):
    hover = HoverTool(
        tooltips=[
            ("Date", "@date{%d %b %Y}"),  # Display the month and year
            ("Net Futures", "@net_futures{0,0}")  # Display the net_futures value with commas
        ],
        formatters={
            '@date': 'datetime'  # Use 'datetime' formatter for the date
        },
        mode='vline'  # Display a vertical line across the plot at the hovered point
    )
    plot.state.add_tools(hover)

# Set the plot options with human-readable y-axis and month formatting for x-axis
net_futures_plot.opts(
    title="Net Futures for Clients Over Time",
    xlabel="Date",
    ylabel="Net Futures",
    width=2000,  # Set a large width to take full width of the screen
    height=800,
    yformatter=NumeralTickFormatter(format="0a"),  # Human-readable format
    xformatter=DatetimeTickFormatter(months="%b %Y"),  # Display month and year
    hooks=[add_hover_tool]  # Use the hook to add the HoverTool
)

# Display the plot
show(hv.render(net_futures_plot))
```
